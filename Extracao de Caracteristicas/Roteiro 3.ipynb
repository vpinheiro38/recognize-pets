```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from skimage import filters, io, feature as ft, color

def toUint8(image):
    return np.array(image, dtype=np.uint8)

def openFile(fileName):
    return open(fileName, 'ab+')

def openHDF5(filename):
    return h5py.File(filename, "a")
```

```python
dataType = ['train']
dirs     = ['eqhist', 'gamma1', 'gamma2', 'gamma3', 'gaussian', 'laplaceg']
petName  = ['cat', 'dog']

for data in dataType:
    dset = openHDF5('data' + data + '.hdf5')
    dsetLBP = dset.create_dataset("dataLBP", (10000, 40*40*6), dtype='f')
    dsetGLCM = dset.create_dataset("dataGLCM", (10000, 40*40*6), dtype='f')
    dsetSobel = dset.create_dataset("dataSobel", (10000, 40*40*6), dtype='f')

    for i in range(0, 2500):
        for pet in petName:
            print(pet + " " + str(i))
            vetLBP   = []
            vetGLCM  = []
            vetSobel = []
            
            for dir in dirs:
                imgOrig = io.imread(data + '/' + pet + 's/'+ pet +'.' + str(i) + '.jpg')
                img = io.imread(data + '/' + pet + 's/dataset40/' + dir + '/'+ pet + '.' + str(i) + '.jpg')

                imgGray  = color.rgb2gray(img)
                imgGray  = imgGray*255.0/imgGray.max()

                radius   = 1
                n_points = 8 * radius

                imgLBP   = ft.local_binary_pattern(imgGray, n_points, radius)
                
                imgSobel = filters.sobel(imgGray)

                contrastraster = np.copy(img)
                contrastraster[:] = 0

                for z in range(img.shape[0]):
                    for k in range(img.shape[1]):
                        #windows needs to fit completely in image
                        if z <3 or k <3:        
                            continue
                        if z > (contrastraster.shape[0] - 1) or k > (contrastraster.shape[0] - 1):
                            continue

                        #Define size of moving window
                        glcm_window = img[z-3: z+1, k-3 : k+1]
                        #Calculate GLCM and textures
                        glcm = ft.greycomatrix(glcm_window, [1], [0],  symmetric = True, normed = True )

                        # Calculate texture and write into raster where moving window is centered
                        contrastraster[z,k] = ft.greycoprops(glcm, 'contrast')
                        glcm = None
                        glcm_window = None
                
                vetLBP   = np.append(vetLBP, imgLBP.flatten())
                vetGLCM  = np.append(vetGLCM, contrastraster.flatten())
                vetSobel = np.append(vetSobel, imgSobel.flatten())
                
            #print(vetLBP)
            #print(vetGLCM)
            #print(vetSobel)
            
            if (pet == 'cat'):
                dsetLBP[2*i] = vetLBP
                dsetGLCM[2*i] = vetGLCM
                dsetSobel[2*i] = vetSobel
            else:
                dsetLBP[2*i+1] = vetLBP
                dsetGLCM[2*i+1] = vetGLCM
                dsetSobel[2*i+1] = vetSobel

dset.close()

print('Roteiro 3 - Processo concluido')
```
